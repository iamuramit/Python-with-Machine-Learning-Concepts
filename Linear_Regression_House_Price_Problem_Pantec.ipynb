```python
size = df['sqft_living'] # input columns
price = df['price'] # output/ target columns

x=np.array(size).reshape(-1,1)
y=np.array(price).reshape(-1,1)

print("size data: ", x[:2])
print("price data: ", y[:2])
```

```python
# we use linear regression + fit() is the training
model = LinearRegression()
model.fit(x,y)
```

```python
#MSE and R value
regression_model_mse = mean_squared_error(x,y)
print("MSE :", math.sqrt(regression_model_mse))
print("R squared value :", model.score(x,y)) #regresstion or root error score
```

```python
#we can get the b values after the model fit
print(model.coef_[0]) #this is the b0
print(model.intercept_[0]) #this is the b1 in our model interception
```

```python
#visualize the dataset with the fitted model
plt.scatter(x,y, color='red')
plt.plot(x, model.predict(x), color='black')
plt.title('Linear Regression')
plt.xlabel("Size")
plt.ylabel('Price')
plt.show()

#predicting the prices
print('Prediction by the model:', model.predict([[2000]]))
```
