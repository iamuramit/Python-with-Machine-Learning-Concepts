## Basic Plots using Seaborn 

### Distance Plot (displot)

```python
import numpy as np
import seaborn as sns

sns.set(style = 'white')
rs = np.random.RandomState(10)
d= rs.normal(size=100)
sns.displot(d,kde=True, color='m')
```

### Line Plot (lineplot) 

```python
sns.set(style='dark')
fmri = sns.load_dataset('fmri')
sns.lineplot(x='timepoint', y='signal', hue='region', style='event', data=fmri)
```

### Linear Model Plot (lmplot) 

```python
import seaborn as sns
sns.set_theme(style = 'ticks')
df = sns.load_dataset("anscombe")
sns.lmplot(x="x", y="y", data=df)
```

```python
import seaborn as sns
sns.set_theme(style="ticks")

# Load the example dataset for Anscombe's quartet
df = sns.load_dataset("anscombe")

# Show the results of a linear regression within each dataset
sns.lmplot(
    data=df, x="x", y="y", col="dataset", hue="dataset",
    col_wrap=2, palette="muted", ci=None,
    height=4, scatter_kws={"s": 50, "alpha": 1}
)
```

## Bar Plot 

```python
sns.set_context('paper')
titanic= sns.load_dataset('titanic')
titanic
```

```python
sns.set_context('paper')
titanic= sns.load_dataset('titanic')
sns.barplot(x='embark_town', y='age', data=titanic, palette='PuRd', ci=None)
plt.legend()
plt.show()
print(titanic.columns)
```

## Bar Chart with Verticle Lines (Verticle Bar Chart) 

```python
titanic= sns.load_dataset('titanic')
sns.barplot(x='sex', 
            y='survived', 
            hue='class', 
            data=titanic, 
            palette='PuRd', 
            order=['male','female'], 
            capsize=0.05, 
            errcolor='gray',
            errwidth=2,
            ci='sd')
plt.legend()
plt.show()
```

## Horizontal Chart 

```python
sns.set_context('paper')
titanic= sns.load_dataset('titanic')
sns.barplot(x='age', y='embark_town', data=titanic, palette='PuRd', orient='h', ci=None) #orient='h' for horizontal
plt.legend()
plt.show()
```

## Count Plot 

```python
sns.set_context('paper')
titanic= sns.load_dataset('titanic')
sns.countplot(x='class', hue='who', data=titanic, palette='magma') #who is a column name in titanic dataset
plt.title('Survivors')
plt.show()
```

## Point Plot

## HeatMap 

```python
flights = sns.load_dataset('flights')
flights = flights.pivot('month', 'year', 'passengers')
print(flights)
sns.heatmap(flights, linewidths=0.5, cmap='YlGnBu')
```

## Facet Grid 

```python
Tips = sns.load_dataset('tips')
G=sns.FacetGrid(Tips, col='time')
G.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
G=sns.FacetGrid(Tips, col='time', row='sex')
G.map(sns.scatterplot, 'total_bill', 'tip')
```
